**Explanation**

First step was to install the OpenAI python library. Once installed I created a project in OpenAI's API platform and created a APi_key. Put some balance in there.

I setup Json, os and openAI libraries. 
Json to convert the results from openAI , os.getenv to fetch the APIkey from the terminal path instead of using the key in code itself, and OpenAI to connect and communicate with model. 

```python
import json
import os
from openai import OpenAI
api_key = os.getenv("OPENAI_API_KEY")
```

Lets define a function based on OpenAI's API documentation (https://developers.openai.com/api/docs/guides/prompt-engineering?lang=python) for responses. I will use this function to fetch the results and print them. this will make the code little sleek.

```python

def openai_prompt(prompt,effort="low"):
    client = OpenAI(api_key=api_key)    
    result = client.responses.create(
         model="gpt-5.1",
         reasoning={ "effort": effort },
         input=(prompt),
            )
    return result.output_text
```

In this first prompt I am creating the Zero shot prompt which will be just my instructions in the prompt and the email.  While writing the prompt I made sure that everything is provided to avoid any hallucinations that model could face. The structure I chose to provide the what to do first, then what should be the output and the specific rules regarding the JSON outputs.This specific step is to avoid hallucinations. At the end, the email is provided which is the input.

```python
# Lets define our Zero prompt for the shoe order email.

prompt = """
Extract all required information from the email below and output ONLY valid JSON. Do not include explanations, 
notes, or commentary.In addition, you'll assign a shoe_price for each shoe, calculate the shoe_subtotal for each shoe, 
and then the grand_total. 

Extract the following fields:

- shoe_agent_name (the person being addressed in the greeting)
- request_type (always "Order")
- customer_name
- customer_street
- customer_city
- customer_state
- customer_zip
- customer_phone
- items: an array where each item contains:
    - shoe_brand
    - shoe_model
    - shoe_quantity
    - shoe_price (assign a reasonable price yourself)
    - shoe_subtotal (quantity × price)
- grand_total (sum of all shoe_subtotals)

Rules:
- Output must be valid JSON.
- shoe_price must be a number.
- shoe_subtotal and grand_total must be correct mathematical calculations.
- Preserve the exact text of shoe brand and model as written in the email.

Email:
Hey Jim,

I'd like to order some new shoes. Please ship the following:

1. Nike Air Jordan I - 2 pair
2. Converse All-Star - 10 pair
3. New Balance 990 - 1 pair
4. Nike Zoom Fly 5 men's red - 2 pair

Please provide sub-totals and grand total cost.

Thanks.

Lance Gentry
123 Main St.
Chelsea, MI 48109
248-229-2229


"""

results = json.loads(openai_prompt(prompt))
print(json.dumps(results, indent=2))
```

As you see above the zero-shot extraction produced valid structured JSON and successfully captured the core entities:
- customer identity and address fields
- item-level attributes
- computed subtotals and grand total
